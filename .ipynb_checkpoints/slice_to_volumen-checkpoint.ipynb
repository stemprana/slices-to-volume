# Slices to volume
Processing and alignment of images obtained utilizing automated confocal microscopy scanning of histology slides with brain slices and creation of a volumetric visualization.


#### Relevant experimental information: 
- Fluorescent proteins of interest are expressed only in one hemisphere and recorded in *green* and *red* channel.
- Fluorescent signal on *blue* channel labels nuclei and is used for registration purposes.
- Low resolution tiles of *blue* channel (referred as **maps**) and high resolution tiles of the three channels (further called **stacks**) were obtained.
- Registration is done using *map* images and applied to corresponding zoomed in *stacks*.

```python
# IMPORTs -----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
import os
import copy
import numpy as np
import json
import napari
import matplotlib.pyplot as plt
import cv2
from skimage.transform import downscale_local_mean
from tifffile import memmap
import slice_to_volume as slc_to_vol
from pathlib import Path
%matplotlib inline
```

To try out code, download input files from [shared folder](https://drive.google.com/drive/folders/14qJdoznqLLFRMbWKD967pK12O5Om-7-L?usp=sharing) and set *save_path* variable to directory where they are saved.
Files have been preprocessed from raw data for convenience,
- **map_info.json,** image specifications for lower magnification *maps*
- **img_map_single.tif,** memory mapped file containing lower magnification *map* images
- **zoomed_info.json,** image specifications for higher magnification *stacks*
- **img_stck_red.tif,** memory mapped file containing higher magnification *stack* images
- **clean_msks_coords.json,** file containing coordinates to clean up *map* images (see below)



```python
# PATHs -----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
save_path = Path('')
```

```python
# Load files specifications ----------------------------------------------------------------------------------    
stack_inf = slc_to_vol.load_json_info('zoomed_info.json',save_path)
map_inf = slc_to_vol.load_json_info('map_info.json',save_path)
```

### 1. Visualize global layout


```python
stg_um,track_bnds_d,map_idxs, stack_idxs = slc_to_vol.map_global_layout(map_inf,stack_inf)
```

Manual definition of order in which coronal brain slices were mounted from anterior to posterior location using previous layout.

```python
# Define anterior to posterior order from display ------------------------------------------
map_seq_ids = [6,7,8,9,13,12,11,10,14,15,16,17,21,20,19,18,22,23,24,25,29,28,27,26]
stck_seq_ids = [30,31,32,33,37,36,35,34,38,39,40,41,45,44,43,42,46,47,48,49,53,52,51,50]
```

#### 1.1 Loading of zoomed-out images and full projection display 

```python
# Load zoomed-out low-resolution images as memory mapped array, images are sorted from anterior to posterior location (dimensions = n_images,y,x) ------------------------
imgs_map = memmap(save_path/'img_map_single.tif')
```

```python
# Assemble full display by positioning images in stage location -----------------------------------------------------
map_proj,map_proj_pxl_size = slc_to_vol.get_map_proj(imgs_map,map_inf,map_idxs,map_seq_ids,track_bnds_d,stg_um)
```

```python
# Display assembly of low resolution images of brain slices -----------------------------------
slc_to_vol.plot_map_imgs(map_proj,stack_inf,stack_idxs,map_proj_pxl_size,track_bnds_d,stg_um)
```

### 2 Perform alignment of zoomed-out images of brain slices

#### Load zoomed-in high-resolution images 

```python
imgs_stck_red = memmap(save_path/'imgs_stck_red.tif')
# Obtain z-average projection of blue channel ---------------------
imgs_stck = np.mean(imgs_stck_red[:,0,...],axis=1)
```

#### 2.1 Detect flipped images and correct

```python
# Flip detection --------------------------------------------------------------------
flip_trk,flip_cc_d,flip_imgs = slc_to_vol.detect_flip(imgs_stck)
```

```python
# Flip correction -----------------------------------------------------------------------------------------------------
imgs_stck_flp,imgs_map_flp  = slc_to_vol.correct_flip(flip_trk,imgs_stck,imgs_map)
```

#### 2.2 Isolate slices layout
Delineate slices on each image to improve performance of intensity based registration.
- File with masks coordinates for each slice is provided - *clean_msks_coords.json*. In order to use it skip next cells and jump to **2.2.2 Load outline coordinates**
- If manually drawing of the outlines is desired continue on next cell, **2.2.1 Display images and outline slices**. 



##### 2.2.1 Display images and outline slices
In order to outline slices execute following code cell and follow these instructions:
- Make sure slider in first slice (*number 0*)
- Add **new shapes layer**
- Modify name to: *slice_0*
- With layer selected click on **Add polygons**
- Draw polygon enclosing brain slice of interest
- Position slider in next slice and repeat instructions, naming each new shapes layer as *slice_(sliceNumber)*

Check following link for information on how to use [napari shapes layer](https://napari.org/stable/howtos/layers/shapes.html)


```python
#Display images in napari to outline shapes -------------------------------------------------------------------
n_vwr = napari.Viewer() # Initialize viewer 
img_lyr = n_vwr.add_image(imgs_map_flp,name='map images flipped',colormap='gray_r') # Add image layer with zoomed out stack
n_vwr.dims.current_step = (0,0,0) # Position slider on first slice
```

Once all slices have been labeled run following cell to retrieve polygon coordinates and save as .json file.      

**WARNING**: Existent files with same name are going to be overwritten

```python
# Retrieve coordinates of masks -------------------------------------------------------------------------
lyrs_name_l = [i.name for i in n_vwr.layers]
map_clean_msks = {}
try:
    for e_s_map_idx in range(len(imgs_map_flp)): 
        assert 'slice_'+str(e_s_map_idx) in lyrs_name_l, f'Mask for slice {e_s_map_idx} has not been found'    
        s_cln_msk = n_vwr.layers['slice_'+str(e_s_map_idx)].data
        assert len(s_cln_msk)==1 and np.all(s_cln_msk[0][:,0]==e_s_map_idx), f'Error found on mask for slice {e_s_map_idx}'
        map_clean_msks['slice_'+str(e_s_map_idx)] = s_cln_msk[0][:,1:].copy()
    # Save coordinates as json ------------------------------------
    map_clean_msks_json = {k:map_clean_msks[k].tolist() for k in map_clean_msks.keys()}        
    with open(os.path.join(save_path,'clean_msks_coords.json'),'w') as clean_msks_out:
        json.dump(map_clean_msks_json,clean_msks_out)
    # --------------------------------------------------------------
except AssertionError as mssg:
    print('File with masks has not been saved')
    print(mssg)
    map_clean_msks = None
```

##### 2.2.2 Load outline coordinates

```python
# Load outline coordinates -------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------------------------------
with open(save_path/'clean_msks_coords.json','r') as json_file_in:    
    msks_crds = json.load(json_file_in)
# --------------------------------------------------------------------------------------------------------------------
```

##### 2.2.3 Apply mask to images

```python
imgs_map_msk,imgs_map_msk_eq,_ = slc_to_vol.get_masked_map(imgs_map_flp,msks_crds,view_imgs=False)
```

#### 2.3 Register brain slices

```python
imgs_map_reg,imgs_map_reg_eq,warp_map =  slc_to_vol.register_map_imgs(imgs_map_msk)
```

### 3. Apply registration to zoomed-in areas and generate volumetric representation
#### 3.1 Transform zoomed-in layout using warp-matrix obtained from intensity based registration

```python
zoomed_transformed_d = slc_to_vol.zoomed_apply_transform(map_seq_ids,stck_seq_ids,map_inf,stack_inf,warp_map,stg_um,flip_trk,imgs_map_msk)
```

#### 3.2 Find spatial overlap between transformed layouts

```python
int_bbox,msks_int_cntr = slc_to_vol.get_zoomed_intersect(zoomed_transformed_d)
```

#### 3.3 Align region of interest of each slice into common reference frame

```python
# Perform alignment on each individual slice -----------------------------------------------------------------------------
slc_to_vol.align_zoomed_in(int_bbox,map_inf,stack_inf,map_seq_ids,stck_seq_ids,imgs_stck_red,zoomed_transformed_d,save_path,flip_trk)
```

```python
# Match dimensions in aligned images ------------------------------------------------------------------------
slc_to_vol.zoomed_in_dims_match(save_path)
```
